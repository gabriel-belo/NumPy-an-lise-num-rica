a= Coeficiente angular

n -> número de elementos

y= Moscow

x= datas

O coeficiente angular da a taxa de crescimento


O coeficiente angular pode ser obtido usando a equação:

â = n * Soma (Xi * Yi) - Soma (Xi) * Soma (Yi)/ n*Soma(Xi²) - (Soma(Xi))²

```python
#aplicando a fórmula
Y= Moscow
x= datas
n= np.size(Moscow)
```

```python
a = (n * np.sum(x*Y)- np.sum(x) * np.sum(Y))/ (n * np.sum(x**2) - np.sum(x)**2)
```

Coeficiente linear
b= Media(Yi) - â * Media(Xi)

```python
b = np.mean(Y) - a * np.mean(x)
```

```python
y= a * x + b
```

```python
np.linalg.norm(Moscow - y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
```

```python
#Fazendo a estimativa para qual deveria ser o valor da maça em certa época
#de acordo com um ponto gerado na reta
plt.plot(datas, Moscow)
plt.plot(x, y)
plt.plot(41.5, 41.5 * a + b, '*r')
```
